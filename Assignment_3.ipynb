```python
# Exercise 1
def BinarySearchRec(lista, start, end, key):
    if start > end:
        return False
    mid = (start + end) // 2
# base case
    if lista[mid] == key:
        return True
# recursive call
    elif lista[mid] > key:
        return BinarySearchRec (lista, start, mid-1, key)
    else:
        return BinarySearchRec (lista, mid+1, end, key)

result1 = BinarySearchRec([-77,-50,-21,-10,-1,0,3,10,21,50,99], 0, 10, 99)
result2 = BinarySearchRec([-77,-50,-21,-10,-1,0,3,10,21,50,99], 0, 10, 100)
print(result1, result2)
```

```python
# Exercise 2
def GenerateCombinations(characters, n, sentence=""):
    # base case
    if n==0:
        print(sentence)
        return
    # recursive call
    for character in characters:
        GenerateCombinations(characters, n-1, sentence + character)

GenerateCombinations(['a','b','c'], 2)
```

```python
# Exercise 3
def addValue(lista, value):
    start = 0
    end = len(lista)
    
    while (start < end):
        mid = (start + end) // 2
        if lista[mid] < value:
            start = mid+1
        else:
            end = mid
    lista.insert(start,value)
    print(lista)

addValue([1,34,56,78,89],77)
addValue([1,34,56,234,789],-99)
addValue([1,34,56,234,789],789)
```

```python
# Exercise 4
items = {
    ("Water", "0000000", 25000),
    ("Cigarettes", "1111111", 75000),
    ("Chocolate", "2222222", 50000),
    ("Ice Cream", "3333333", 40000),
    ("Dairy Product", "4444444", 100000),
    ("Bread", "5555555", 55000),
    ("Vegetables", "6666666", 20000),
    ("Fruits", "7777777", 30000)
}

def get_item_details(barcode):
    for name, item_barcode, price in items:
        if item_barcode == barcode:
            return name, price
    return None, None

def add_item_to_receipt(receipt):
    item_barcode = input("What is the barcode of the item the client wants to buy? ")
    quantity = int(input("How many items does the client want to buy? "))
    
    item_name, price_per_item = get_item_details(item_barcode)
    
    if item_name:
        for index, (name, barcode, qty, price) in enumerate(receipt):
            if barcode == item_barcode:
                receipt[index] = (name, barcode, qty + quantity, price)
                return
        receipt.append((item_name, item_barcode, quantity, price_per_item))
    else:
        print("Item not found.")

def format_price(price):
    return f"{price // 1000}k LBP"

def print_receipt(receipt):
    total_price = sum(quantity * price_per_item for _, _, quantity, price_per_item in receipt)
    for name, barcode, quantity, price_per_item in receipt:
        print(f"Item: {name}, Barcode: {barcode}, Quantity: {quantity}, Price per item: {format_price(price_per_item)}, Total: {format_price(quantity * price_per_item)}")

    print(f"The total cost is: {format_price(total_price)}")

answer1 = input("Hi 3ammo l dekkanje, would you like to start a new receipt? yes/no ")

while answer1.lower() == "yes":
    receipt = []
    
    add_item_to_receipt(receipt)

    answer2 = input("Would you like to add another item to the list? yes/no ")

    while answer2.lower() == "yes":
        add_item_to_receipt(receipt)
        answer2 = input("Would you like to add another item to the list? yes/no ")

    if answer2.lower() == "no":
        print_receipt(receipt)
    else:
        print("Please 3ammo answer with either 'yes' or 'no'.")

    answer1 = input("Would you like to start a new receipt? yes/no ")

if answer1.lower() == "no":
    print("Have a nice day 3ammo!")
else:
    print("Please 3ammo answer with either 'yes' or 'no'.")
```
