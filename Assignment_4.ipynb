```python
# Stacks and Queues

#Exercise 1: Palindrome

def testPalindrome(sentence):
    filtered_sentence = "".join(char.lower() for char in sentence if char.isalnum())
    stack = []
    for i in range (len(filtered_sentence)):
        stack.append(filtered_sentence[i])
    for j in range (len(filtered_sentence)):
        if (stack.pop() == filtered_sentence[j]):
            return True
        else:
            return False

print(testPalindrome("Mom"))
print(testPalindrome("Dad"))
print(testPalindrome("Never odd or even"))
print(testPalindrome("no palindrome"))
print(testPalindrome("7alab like, ekil bala7"))
print(testPalindrome("Ana tor, Rotana"))
```

```python
# Exercise 2

def testBalance(expression):
    open_brackets = "({["
    closed_brackets=")}]"
    matching_bracket = {')': '(', '}': '{', ']': '['}
    stack = []
    for element in expression:
        if element in open_brackets:
            stack.append(element)
        elif element in closed_brackets:
            if not stack or stack[-1] != matching_bracket[element]:
                return False
            stack.pop()
    return True

print(testBalance("(1+2)-3*[41+6]"))
print(testBalance("(1+2)-3*[41+6}"))
print(testBalance("(1+2)-3*41+6]"))
print(testBalance("(1+2)-3*]41+6["))
print(testBalance("(1+[2-3]*{41+6}"))
```

```python
# Stack

def MIB(code):
    stack = []
    decode = []
    for element in code:
        if (element.isalpha() or element.isspace()):
            stack.append(element)
        elif (element == "*"):
            if stack:
                decode.append(stack.pop())
            else:
                decode.append("*")
    while stack:
        decode.append(stack.pop())
        message = ''.join(decode)
       
    return message

print(MIB("SIVLE ****** DAED TNSI ***"))
```

```python
# Linked Lists - removing a node at a given location

class Node:
    def __init__(self, info):
        self.info = info
        self.next = None

class LL:
    def __init__(self):
        self.head = None
    
    def deleteAtLocation(self, position):
        if self.head == None:
            print("The linked list is empty")
            return
        
        if position == 0:
            self.head = self.head.next
            return
        if position < 0:
            print("out of bounds")
            return
        
        my_node = self.head
        previous = None
        step = 0
        
        while my_node and step < position:
            previous = my_node
            my_node = my_node.next
            step += 1
        
        if my_node == None:
            print("out of bounds")
            return
        
        previous.next = my_node.next
        
    def printLL(self):
        temp = self.head
        while temp != None:
            print(temp.info," -> ", end=" ")
            temp = temp.next
        print()

linked_list = LL()
linked_list.head = Node(12)
second = Node(56)
third = Node(76)
fourth = Node(11)
fifth = Node(0)
linked_list.head.next = second
second.next = third
third.next = fourth
fourth.next = fifth

print("main linked list:")
linked_list.printLL()

linked_list.deleteAtLocation(0)
print("linked list after deleting node at position 0")
linked_list.printLL()

linked_list.deleteAtLocation(3)
print("linked list after deleting node at position 3")
linked_list.printLL()
```
